```python
import random

# Deck of cards - 52 cards in total (13 cards in each of 4 suits: Spades, Hearts, Diamonds, Clubs)
suits = ['Spades', 'Hearts', 'Diamonds', 'Clubs']
ranks = ['2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 'Queen', 'King', 'Ace']

# Creating the deck
deck = [(rank, suit) for suit in suits for rank in ranks]

# Black cards are in the suits of Spades and Clubs
black_cards = [(rank, suit) for rank, suit in deck if suit in ['Spades', 'Clubs']]

# Finding the Black Kings
black_kings = [(rank, suit) for rank, suit in black_cards if rank == 'King']

# Output results
print(f"Total black cards: {len(black_cards)}")
print(f"Black Kings: {black_kings}")

# Simulate randomly picking a card from the black cards
picked_card = random.choice(black_cards)

print(f"Picked card: {picked_card}")

# Check if the picked card is a black king
if picked_card in black_kings:
    print("The picked card is a black King.")
else:
    print("The picked card is not a black King.")



```

```python
# Total number of cards in the deck
total_cards = 52

# Total number of black cards (Spades and Clubs)
total_black_cards = 26

# Number of black kings (King of Spades and King of Clubs)
total_black_kings = 2

# P(A) = Probability of drawing a black king
P_A = total_black_kings / total_cards

# P(B|A) = Probability of drawing a black card given it's a black king
# Since both black kings are also black cards, this is 1 (certainty)
P_B_given_A = 1

# P(B) = Probability of drawing a black card
P_B = total_black_cards / total_cards

# Apply Bayes' Theorem: P(A|B) = (P(B|A) * P(A)) / P(B)
P_A_given_B = (P_B_given_A * P_A) / P_B

# Output the results
print(f"P(A|B) = {P_A_given_B:.4f} (Probability of drawing a Black King given it's a Black Card)")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Generating a random dataset with a normal distribution (mean = 50, std = 5, size = 1000)
data = np.random.normal(50, 5, 1000)

# Calculating mean, median, and mode
mean_value = np.mean(data)
median_value = np.median(data)
mode_value = stats.mode(data, keepdims=True)[0][0]

# Plotting the density using seaborn
plt.figure(figsize=(10, 6))
sns.kdeplot(data, color='blue', shade=True)

# Adding mean, median, and mode to the plot
plt.axvline(mean_value, color='red', linestyle='--', label=f'Mean: {mean_value:.2f}')
plt.axvline(median_value, color='green', linestyle='-', label=f'Median: {median_value:.2f}')
plt.axvline(mode_value, color='orange', linestyle='-', label=f'Mode: {mode_value:.2f}')

# Graph labels and legend
plt.title('Density Plot with Mean, Median, and Mode')
plt.legend()
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()

# Display the calculated values
mean_value, median_value, mode_value
```

```python
# Given probabilities
P_S = 0.2  # Probability that any email is spam
P_W_given_S = 0.4  # Probability of "discount" appearing in spam emails
P_W_given_not_S = 0.1  # Probability of "discount" appearing in non-spam emails
P_not_S = 1 - P_S  # Probability that an email is not spam

# Calculate P(W) using the total probability formula
P_W = (P_W_given_S * P_S) + (P_W_given_not_S * P_not_S)

# Apply Bayes' Theorem to calculate P(S|W)
P_S_given_W = (P_W_given_S * P_S) / P_W

# Output the result
print(f"The probability that the email is spam given that it contains 'discount' is {P_S_given_W:.2f}")
```
