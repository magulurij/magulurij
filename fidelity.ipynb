```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from textblob import TextBlob

# URL of the page to scrape
url = 'https://fidelity.app.link/e/DMtUN9ndyxb'

# Send a GET request to the webpage
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the page content with BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract comments or reviews (you'll need to inspect the structure of the page)
    # In this example, let's assume comments are in <p> tags
    comments = soup.find_all('p')

    # List to hold the scraped data
    comment_data = []

    # Loop through each comment and analyze sentiment
    for comment in comments:
        comment_text = comment.text.strip()  # Get the comment text

        # Perform sentiment analysis using TextBlob
        sentiment = TextBlob(comment_text).sentiment.polarity

        # Classify the sentiment as Positive, Negative, or Neutral
        if sentiment > 0:
            sentiment_label = 'Positive'
        elif sentiment < 0:
            sentiment_label = 'Negative'
        else:
            sentiment_label = 'Neutral'

        # Append the data to the list
        comment_data.append({
            'Comment': comment_text,
            'Sentiment': sentiment_label
        })

    # Convert the list to a Pandas DataFrame
    df = pd.DataFrame(comment_data)

    # Display the DataFrame
    print(df)

    # Optionally, save the DataFrame to a CSV file
    df.to_csv('comments_sentiment.csv', index=False)

    print("Comments and sentiment saved to comments_sentiment.csv")

else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
from bs4 import BeautifulSoup
from textblob import TextBlob

# Sample HTML containing YouTube comments (a simplified example for illustration)
html_content = """
<div class="comments">
    <div class="comment">
        <span class="comment-text">This is a fantastic video, very informative!</span>
    </div>
    <div class="comment">
        <span class="comment-text">I didn't like the presentation style, very boring.</span>
    </div>
    <div class="comment">
        <span class="comment-text">It was okay, nothing too special or exciting.</span>
    </div>
    <div class="comment">
        <span class="comment-text">Great work! Looking forward to more content like this.</span>
    </div>
    <div class="comment">
        <span class="comment-text">The information was inaccurate, very disappointing.</span>
    </div>
</div>
"""

# Step 1: Parse the HTML content with BeautifulSoup
soup = BeautifulSoup(html_content, 'lxml')

# Step 2: Extract all comments
comments_tags = soup.find_all('span', class_='comment-text')
comments = [comment.text.strip() for comment in comments_tags]

# Step 3: Analyze sentiment for each comment
def analyze_sentiments(comments):
    sentiment_results = []

    for comment in comments:
        # Perform sentiment analysis using TextBlob
        analysis = TextBlob(comment)

        # Determine sentiment polarity
        if analysis.sentiment.polarity > 0:
            sentiment = "Positive"
        elif analysis.sentiment.polarity < 0:
            sentiment = "Negative"
        else:
            sentiment = "Neutral"

        # Store the result as a dictionary
        sentiment_results.append({
            "comment": comment,
            "sentiment": sentiment,
            "polarity": analysis.sentiment.polarity
        })

    return sentiment_results

# Get the sentiment analysis results
sentiment_results = analyze_sentiments(comments)

# Step 4: Print the comments along with their sentiment
for result in sentiment_results:
    print(f"Comment: {result['comment']}")
    print(f"Sentiment: {result['sentiment']} (Polarity: {result['polarity']:.2f})")
    print("=" * 80)
```
